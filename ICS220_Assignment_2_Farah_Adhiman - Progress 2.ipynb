```python
class Guest:

    # Initializing the attributes for the Guest class (private)
    def __init__(self, nameofguest, emailguest, guestphonenumber, guestid, guestloyalty):
        self.__nameofguest = nameofguest
        self.__emailguest = emailguest
        self.__guestphonenumber = guestphonenumber
        self.__guestid = guestid
        self.__guestloyalty = guestloyalty

    # Getter methods
    def get_nameofguest(self):
        return self.__nameofguest

    def get_emailguest(self):
        return self.__emailguest

    def get_guestphonenumber(self):
        return self.__guestphonenumber

    def get_guestid(self):
        return self.__guestid

    def get_guestloyalty(self):
        return self.__guestloyalty

    # Setter methods
    def set_nameofguest(self, nameofguest):
        self.__nameofguest = nameofguest

    def set_emailguest(self, emailguest):
        self.__emailguest = emailguest

    def set_guestphonenumber(self, guestphonenumber):
        self.__guestphonenumber = guestphonenumber

    def set_guestloyalty(self, guestloyalty):
        self.__guestloyalty = guestloyalty

    # Guest actions
    def guestcreateaccount(self):
        return "The guest account for", self.__nameofguest, "has been created!"

    def updateguestprofile(self):
        return "The guest profile for guest ID:", self.__guestid, "has been updated."

    def guestreservation(self):
        return "Displaying reservation history for guest with ID:", self.__guestid, "."

    # String representation
    def __str__(self):
        return f"Guest ID: {self.__guestid} - Name: {self.__nameofguest} - Email: {self.__emailguest} - Phone: {self.__guestphonenumber} - Loyalty: {self.__guestloyalty}"


class Booking:

    # Initializing the attributes for the Booking class (private)
    def __init__(self, guestid, roomnum, room_booking_id, statusofbooking, dateofcheckin, dateofcheckout):
        self.__guestid = guestid
        self.__roomnum = roomnum
        self.__room_booking_id = room_booking_id
        self.__statusofbooking = statusofbooking
        self.__dateofcheckin = dateofcheckin
        self.__dateofcheckout = dateofcheckout

    # Getter methods
    def get_guestid(self):
        return self.__guestid

    def get_roomnum(self):
        return self.__roomnum

    def get_room_booking_id(self):
        return self.__room_booking_id

    def get_statusofbooking(self):
        return self.__statusofbooking

    def get_dateofcheckin(self):
        return self.__dateofcheckin

    def get_dateofcheckout(self):
        return self.__dateofcheckout

    # Setter method
    def set_statusofbooking(self, status):
        self.__statusofbooking = status

    # Guest actions
    def checkavailability(self):
        return "Searching for rooms that are available..."

    def makereservation(self):
        return "Booking for guest with ID:", self.__guestid, "in room", self.__roomnum, "has been made."

    def confirming(self):
        return "Booking confirmation has been sent to guest owners of room:", self.__roomnum, "via Email."

    def cancel_reservation(self):
        self.set_statusofbooking("Cancelled")
        return "Booking with ID:", self.__room_booking_id, "has been cancelled."

    # String representation
    def __str__(self):
        return f"Booking ID: {self.__room_booking_id} - Guest ID: {self.__guestid} - Guest room: {self.__roomnum} - Booking status: {self.__statusofbooking} - Guest check-in: {self.__dateofcheckin} - Guest check-out: {self.__dateofcheckout}"


class Payment:

    # Initializing the attributes for the Payment class (private)
    def __init__(self, invoiceid, payment_id, methodofpayment, total, statusofpayment):
        self.__invoiceid = invoiceid
        self.__payment_id = payment_id
        self.__methodofpayment = methodofpayment
        self.__total = total
        self.__statusofpayment = statusofpayment

    # Getter methods
    def get_invoiceid(self):
        return self.__invoiceid

    def get_payment_id(self):
        return self.__payment_id

    def get_methodofpayment(self):
        return self.__methodofpayment

    def get_total(self):
        return self.__total

    def get_statusofpayment(self):
        return self.__statusofpayment

    # Setter method
    def set_statusofpayment(self, status):
        self.__statusofpayment = status

    # Guest actions
    def checkpaymentstatus(self):
        return "Payment status for payment with ID", self.__payment_id, "is", self.__statusofpayment, "."

    def processofpayment(self):
        return "Processing payment with ID", self.__payment_id, "using", self.__methodofpayment, "."

    # String representation
    def __str__(self):
        return f"Payment ID: {self.__payment_id} - Invoice ID: {self.__invoiceid} - Total payment amount: {self.__total} - Payment status: {self.__statusofpayment}"

class LoyaltyProgram:

    # Initializing the attributes for the LoyaltyProgram class (private)
    def __init__(self, guestid, loyaltypoints, rewardsstatus, requesttype):
        self.__guestid = guestid
        self.__loyaltypoints = loyaltypoints
        self.__rewardsstatus = rewardsstatus
        self.__requesttype = requesttype

    # Getter methods
    def get_guestid(self):
        return self.__guestid

    def get_loyaltypoints(self):
        return self.__loyaltypoints

    def get_rewardsstatus(self):
        return self.__rewardsstatus

    def get_requesttype(self):
        return self.__requesttype

    # Setter methods
    def set_loyaltypoints(self, points):
        self.__loyaltypoints = points

    def set_rewardsstatus(self, status):
        self.__rewardsstatus = status

    def set_requesttype(self, requesttype):
        self.__requesttype = requesttype

    # Guest actions
    def checkrewards_and_status(self):
        return "Guest", self.__guestid, "has", self.__loyaltypoints, "points and reward status is", self.__rewardsstatus, "."

    def redeem(self):
        return "Redeeming loyalty rewards for guest", self.__guestid, "."

    def addmorepoints(self):
        return "Loyalty points have been added for guest with ID:", self.__guestid, "."

    # String representation
    def __str__(self):
        return f"Guest ID: {self.__guestid} - Loyalty points: {self.__loyaltypoints} - Reward status: {self.__rewardsstatus} - Request type: {self.__requesttype}"
```
